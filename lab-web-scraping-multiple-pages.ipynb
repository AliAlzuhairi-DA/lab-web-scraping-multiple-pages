```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get('https://en.wikipedia.org/wiki/Wikipedia:WikiProject_Albums/500')
r.status_code
```

```python
html = r.content
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
df = pd.read_html(soup.find_all('table', attrs={'class': "wikitable plainrowheaders sortable"})[0].prettify())[0]
```

```python
df
```

```python
df = df.drop(['#', 'Major contributors'], axis=1)
```

```python
df
```

```python
df.to_csv('500_Greatest_Albums_of_All_Time.csv', index=False)
```
